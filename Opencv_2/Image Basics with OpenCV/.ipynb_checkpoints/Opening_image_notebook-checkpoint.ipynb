```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img = cv2.imread("image.jpg")
```

```python
type(img)
```

```python
img.shape
```

```python
plt.imshow(img)
```

```python
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img)
```

```python
img_gray = cv2.imread("image.jpg",cv2.IMREAD_GRAYSCALE)
plt.imshow(img_gray,cmap="gray")
```

```python
img_gray.shape
```

```python
img_gray = cv2.resize(img_gray,(100,100))
```

```python
plt.imshow(img_gray,cmap="gray")
```

```python
img_gray.shape
```

```python
img_gray = cv2.flip(img_gray,0)
plt.imshow(img_gray,cmap="gray")
```

```python
cv2.imwrite("gray_image.jpg",img_gray)
```

```python
fig = plt.figure(figsize=(2,2))
ax = fig.add_subplot(111)
ax.imshow(img_gray)
```
