```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread("cube.jpg",cv2.IMREAD_GRAYSCALE)
plt.imshow(img,cmap="gray")
```

```python
gray = np.float32(img)
```

```python
dst = cv2.cornerHarris(gray,blockSize=2,ksize=3,k=0.04)
```

```python
dst = cv2.dilate(dst,None)
```

```python
img[dst>0.01*dst.max()]=(255)
```

```python
plt.imshow(img,cmap="gray")
```

```python
img1 = cv2.imread("cube.jpg")
img1 = cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
gray= cv2.cvtColor(img1,cv2.COLOR_BGR2GRAY)
plt.imshow(img1)
```

```python
corners = cv2.goodFeaturesToTrack(gray,30,0.01,10)
```

```python
corners = np.int0(corners)
corners
```

```python
for i in corners:
    x,y = i.ravel()
    cv2.circle(img1,(x,y),5,(255,0,0),-1)
    print(x,y)
```

```python
plt.imshow(img1)
```
