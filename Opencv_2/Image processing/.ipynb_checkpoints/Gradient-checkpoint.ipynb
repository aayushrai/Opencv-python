```python
import cv2
import matplotlib.pyplot as plt
```

```python
img1 = cv2.imread("g.jpg")
img1 = cv2.cvtColor(img1,cv2.COLOR_BGR2GRAY)
plt.imshow(img1)
```

```python
sobelx = cv2.Sobel(img1,cv2.CV_64F,1,0,ksize=5)   #vertical lines
```

```python
plt.imshow(sobelx)
```

```python
sobelx = cv2.Sobel(img1,cv2.CV_64F,0,1,ksize=5)   #horizontal line
plt.imshow(sobelx)
```

```python
lap = cv2.Laplacian(img1,cv2.CV_64F)
```

```python
fig = plt.figure(figsize=(20,20))
ax=fig.add_subplot(111)
ax.imshow(lap,cmap="gray")
```
