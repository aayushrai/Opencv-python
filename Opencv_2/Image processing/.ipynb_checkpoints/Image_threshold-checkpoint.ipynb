```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('rafale.jpeg',0)
plt.imshow(img,cmap='gray')
```

```python
ret,thres = cv2.threshold(img,125,255,cv2.THRESH_BINARY)
```

```python
plt.imshow(thres,cmap="gray")
```

```python
ret,thres = cv2.threshold(img,125,255,cv2.THRESH_BINARY_INV)
plt.imshow(thres,cmap="gray")
```

```python
thres = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,11,8)
plt.imshow(thres,cmap="gray")
```
