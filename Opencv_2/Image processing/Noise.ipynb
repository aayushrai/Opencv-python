```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = np.zeros([512,512],np.uint8)
cv2.putText(img,"ABCD",(50,300),cv2.FONT_HERSHEY_SIMPLEX,5,(255,255,255),30)
plt.imshow(img,cmap="gray")
```

```python
kernel = np.ones((4,4),dtype=np.uint8)
```

```python
er=cv2.erode(img,kernel,iterations=2)
```

```python
plt.imshow(er,cmap="gray")
```

```python
noise = np.random.randint(low=0,high=2,size=(512,512))
```

```python
plt.imshow(noise,cmap="gray")
```

```python
noise = noise*255
```

```python
noise_img = noise + img
noise_img
```

```python
plt.imshow(noise_img,cmap='gray')
```
